```python
import pandas as pd
```

```python
df = pd.read_csv("/content/dataset_Facebook.csv", delimiter=";")
```

```python
df
```

```python
df.describe()
```

#1)Create subset of dataset

```python
#loc[0:10] = it will take values strat from index 0 to 10
#iloc[0:10] = it will take value start from index 0 to 9
df1 = df[['Type', 'Category', 'like', 'share']].iloc[0:10]
df1
```

```python
df2 = df[df['like'] > 1500]
df2
```

```python
df3 = df[['Type', 'Category', 'comment']].loc[21:50]
df3
```

#2) Merge Dataset

```python
merged_df = pd.merge(df1, df2, on = ['Type', 'Category'], how = 'inner')
merged_df
```

#3) Sort Data

```python
df4 = df[['Page total likes', 'comment', 'share']]
df4
```

```python
df4.sort_values(by = 'Page total likes', ascending=True)
```

#4) Transpose data

```python
#Method 1: df.transpose()
merged_df.transpose()
```

```python
#Method 2: df.T
merged_df.T
```

```python
transpose_sub = df[['Type', 'Category']].T
transpose_sub
```
